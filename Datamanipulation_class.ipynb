```python
# Sometimes you want to calculate a new field which results from complecc calculations derived from one or more
# fields

# we could do a for loop
def complex_function(sales,profit):
    return sales-profit

new_list = []
for index, row in data.iterrows():
    new_list.append(complex_function(row['Sales'],row['Profit']))
                   
```

```python
# Why is the above method not good in the contaxt of big data
new_list
```

```python
# In the context of big data we would not want to load all the rows into the memory to theen do a for loop

# We use the concept of apply

# We want to extract the year component from the date

def complex_function(order_date):
    try:
        result = str(order_date)[0:4]
    except:
        result = None
        
    return result
    
# After defining the function alays test on 1 row
complex_function('2018-02-07')
```

```python
# Now we want to apply the function above row by row in a pandas manner
```

```python
data['Year_of_order'] = data['Order Date'].apply(complex_function)
```

```python
data.head(0)
```

```python
# more complex usecase where our 'comple_function' function receive multiple inputs

def even_more_complex_function(sales, profit, multiplier):
    return multiplier*(sales-profit)

data['new_column'] = data.apply(lambda df : even_more_complex_function(df['Sales'],df['Profit'], 3), axis=1)
```
