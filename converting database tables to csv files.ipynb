```python
import pandas as pd
import sqlite3
# establish database connection
conn = sqlite3.connect('database.sqlite')
c = conn.cursor()
```

```python
# save sqlite table in a DataFrame
df = pd.read_sql('SELECT * from Player', conn)
# write DataFrame to CSV file
df.to_csv('Player.csv', index = False)
```
